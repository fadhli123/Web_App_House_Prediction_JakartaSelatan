### Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,10)
```

### Data Reading

```python
df = pd.read_csv('../Data/HARGA RUMAH JAKSEL ALL.csv')
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

### Data Cleaning

```python
df['HARGA PER METER'] = df['HARGA'] / df['LB']
df['LB PER RUANG'] = df['LB'] / (df['KT'] + df['KM'])
pd.set_option('display.float_format', lambda x: '%.0f' % x)
df['HARGA PER METER'] = df['HARGA PER METER'].astype(np.int64)

df.head(20)
```

```python
df.info()
df.describe()
```

```python
# find outlier in LB PER RUANG
df[df['LB PER RUANG']<20].head(30)
```

```python
#remove data based on ruang
df_cleaned = df[~(df['LB PER RUANG']<10)]
df_cleaned = df[~(df['KT']>10)] # if KT >10 it is boarding house and not house
df_cleaned.reset_index()
df_cleaned.describe()
```

```python
df_cleaned.info()
```

```python
# find outlier in HARGA PER METER
df_cleaned['HARGA PER METER'].describe()
```

```python
mean = np.mean(df_cleaned['HARGA PER METER'])
std = np.std(df_cleaned['HARGA PER METER'])
df_cleaned = df_cleaned[(df_cleaned['HARGA PER METER']>=(mean-std)) & (df_cleaned['HARGA PER METER']<=(mean+std))]
df_cleaned.info()
```

```python
df_cleaned.describe()
```

### Model Building

##### Defining Features & Target

```python
df_model = df_cleaned.drop(['HARGA PER METER', 'LB PER RUANG','KOTA'], axis='columns')
df_model.reset_index(drop=True, inplace=True)
```

```python
dummies = pd.get_dummies(df_model['GARASI'])
dummies
```

```python
df_model = pd.concat([df_model,dummies],axis='columns')
df_model = df_model.drop(['GARASI'],axis='columns')
df_model
```

```python
# Features
X = df_model.drop(columns='HARGA')
X.head()
```

```python
y = df_model['HARGA']
y.head()
```

##### ML Training

```python
# Libraries
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Splitting
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=10)
```

```python
# Linear Regression
lr = LinearRegression(normalize=True)
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
# Cross Validation
cv = ShuffleSplit(n_splits=5, test_size=0.25, random_state=0)
cross_val_score(LinearRegression(), X,y, cv=cv)
```

```python
# defining best algorithms
def find_best_model(X,y):
    algos = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {
                'normalize' : [True,False]
            }
        },
        'lasso' : {
            'model' : Lasso(),
            'params' :{
                'alpha' :[1,2],
                'selection' : ['random','cyclic']
            }
        },
        'decision_tree' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'criterion' : ['mse','friedman_mse'],
                'splitter' : ['best','random']
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.25, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model', 'best_score', 'best_params'])

find_best_model(X,y)
```

Each of algoritmhs can result best score, I'll proceeds with linear regression

```python
# make prediction
def predict_price(LT,LB,KT,KM,garasi):  

    a = np.zeros(len(X.columns))
    a[0] = LT
    a[1] = LB
    a[2] = KT
    a[3] = KM
    if garasi == 'ADA':
        a[4] = 1
        a[5] = 0
    elif garasi == 'TIDAK ADA':
        a[4] = 0
        a[5] = 1

    price = lr.predict([a])[0].astype(np.int64)
    price = round(price, -7)

    return print(f'{price:,}')
```

```python
predict_price(36,36,1,1,'ADA')
```

```python
predict_price(430,420,7,4,'ADA')
```

```python
predict_price(250,250,4,1,'TIDAK ADA')
```
